```python
import numpy as np 
import matplotlib.pyplot as plt

mean_photons = 100
sensor_row = np.random.poisson(mean_photons, size=200)

plt.plot(sensor_row)
plt.xlabel('Pixel')
plt.ylabel('Photon Count')
plt.title('Photon Counts Across Sensor Row')
plt.show()
```

```python
plt.figure()
counts, bins, _ = plt.hist(
    sensor_row,
    bins=30,
    density=True,
    edgecolor='yellow',
    alpha=0.7,
    label='Sample'
)

mean = mean_photons
std = np.sqrt(mean_photons)
x = np.linspace(bins[0], bins[-1], 400)
normal_pdf = (1.0 / (std * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mean) / std) ** 2)

plt.plot(x, normal_pdf, color='black', linewidth=2, label='Normal approx')
plt.axvline(mean, color='red', linestyle='--', linewidth=2, label='Mean (100)')
plt.xlabel('Photon Count')
plt.ylabel('Density')
plt.title('Histogram of Photon Counts')
plt.legend()
plt.show()
```
